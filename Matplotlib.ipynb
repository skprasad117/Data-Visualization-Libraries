# 3. Object Oriented Interface

````python
# creating figure objects and call methods or attributes of that object helps in better dealing with multiple plots 
````

````python
import matplotlib.pyplot as plt
import numpy as np
import math

fig = plt.figure() # creating figure instance to provide empty canvas

# adding axes to figure. this method requires list of 4 elements corresponding to left, bottom, width and height of figure.
ax = fig.add_axes([0,0,1,1]) # each number between 0 and 1

# setting labels
ax.set_title("sine wave")
ax.set_xlabel("angle")
ax.set_ylabel('sine')

# invoking plot method
ax.plot(x,y)
plt.show()
````

## subplot2grid()
- more flexiblity in creating an axes object at a specific location of the grid
```
plt.subplot2grid(shape, location, rowspan, colspan)
```

````python
a1 = plt.subplot2grid((3,3),(0,0),colspan =2)
a2 =  plt.subplot2grid((3,3),(0,2), rowspan=3)
a3 = plt.subplot2grid((3,3),(1,0),rowspan=2,colspan=2)
x = np.arange(1,10)

a2.plot(x,x*x)
a2.set_title('square')

a1.plot(x,np.exp(x))
a1.set_title('exp')

a3.plot(x,np.log(x))
a3.set_title('log')

plt.tight_layout()
plt.show()
````

# grid()

- Turns visiblity of grid on or off

````python
fig,axes = plt.subplots(1,3, figsize =(12,4))
x = np.arange(1,11)

axes[0].plot(x, x**3, 'g', lw=2)
axes[0].grid(True)
axes[0].set_title('default grid')

axes[1].plot(x, np.exp(x), 'r')
axes[1].grid(color = 'b', ls = '-.',lw=0.25)
axes[1].set_title('custom grid')

axes[2].plot(x, x)
axes[2].grid(False)
axes[2].set_title('no grid')

fig.tight_layout()
plt.show()
````
